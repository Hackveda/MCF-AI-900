```python
import nltk
nltk.download('book')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import io
spam = pd.read_csv(io.BytesIO(uploaded['spam.csv']),encoding='ISO-8859-1')
```

```python
spam
```

```python
spam.isnull().sum()
```

```python
x = spam[['v2']]
```

```python
x
x['v2'] = x['v2'].map(lambda x: x.lower())
x
```

```python
from nltk import word_tokenize, sent_tokenize
x['tokenized'] = x['v2'].apply(word_tokenize)
```

```python
x
```

```python
len(x)
```

```python
x = x[['tokenized']]
```

```python
x
```

```python
list = []
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
for i in range(0,len(x)):
    list.append([lemmatizer.lemmatize(text) for text in x['tokenized'][i]])
```

```python
list
```

```python
x['lem'] = list
```

```python
x
```

```python
y = spam[['v1']]
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
y['v1'] = encoder.fit_transform(y[['v1']])
y['v1']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x['lem'],y['v1'], test_size=0.2,random_state=53)
```

```python
X_train=X_train.apply(lambda x: ' '.join(x))
X_test=X_test.apply(lambda x: ' '.join(x))
```

```python
X_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=1000)
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)
```

```python
vectorizer.vocabulary_
```

```python
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
prediction_countV = model.predict(X_test_count)
accuracy_score(y_test,prediction_countV)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df =0.0,max_df=0.9)
X_train_tf = vectorizer.fit_transform(X_train)
X_test_tf = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_tf,y_train)
prediction_countV = model.predict(X_test_tf)
accuracy_score(y_test,prediction_countV)
```
